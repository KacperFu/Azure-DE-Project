# Create Fact Sales

```python
df_silver = spark.sql('''
select * from parquet.`abfss://silver@contosoprojectstorage.dfs.core.windows.net/contoso_sales`
'''
)
```

# Read all dimensions

```python
df_customer = spark.sql("Select * from contoso_catalog.gold.dim_customers")
df_product = spark.sql("Select * from contoso_catalog.gold.dim_prod")
df_category = spark.sql("Select * from contoso_catalog.gold.dim_categ")
df_subcategory = spark.sql("Select * from contoso_catalog.gold.dim_sub")
```

# Join all tables

```python
df_fact = df_silver.join(df_customer,df_silver['customer_key'] == df_customer['customer_key'], how='left')\
                    .join(df_product,df_silver['product_key'] == df_product['product_key'], how='left')\
                        .join(df_category,df_silver['category_key'] == df_category['category_key'], how='left')\
                            .join(df_subcategory,df_silver['subcategory_key'] == df_subcategory['subcategory_key'], how='left')\
                            .select(df_silver['uid'],
                                    df_silver['quantity'],
                                    df_silver['order_date'],
                                    df_silver['delivery_date'],
                                    df_silver['store_key'],
                                    df_silver['currency_code'],
                                    df_silver['exchange_rate'],
                                    df_customer['customer_id'],
                                    df_product['product_id'],
                                    df_category['category_id'],
                                    df_subcategory['subcategory_id'])
```

```python
display(df_fact)
```

# Writing Fact Table

```python
df_fact.write.format('delta')\
          .mode('Overwrite')\
            .option('path','abfss://gold@contosoprojectstorage.dfs.core.windows.net/fact_sales')\
              .saveAsTable('contoso_catalog.gold.fact_sales')
```
